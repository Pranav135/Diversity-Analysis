#### Data Understanding:



#### Let us begin by importing the necessary libraries. For basic Exploratory Data Analysis, we will need 'matplotlib', 'numpy' and 'pandas'. We will also be using 'seaborn' for statistical visualization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### The HR dataset is taken from kaggle.com. We will Import the dataset in 'csv' format as it is and store it in a variable 'hr'. We can then see basic information using 'info'.

```python
hr = pd.read_csv('C:\\Users\\lenovo\\Desktop\\COLLEGE\\Christ\\Trimester 4\\Programming with Python\\HR Dataset.csv')
hr.info()
```

#### Let us see what the dataset holds by listing the first 5 observations:

```python
hr.head(5)
```

#### Let us determine the dimensions of the dataset using 'shape':

```python
hr.shape
```

#### Now, we can see basic statistical inferences from the dataset using 'describe':

```python
hr.describe().transpose()
```

#### Let us check for missing values in the dataset:

```python
hr.isna().sum()
```

```python
hr.isnull().values.any()
```

#### We can notice that there are quite a few missing values. In case we proceed to modify the null values by either neglecting them or removing them, there might be a significant increase in bias of the analysis. Since the aim of the project is to determine diversity and ethnic distributions, this may prove to work as a disadvantage. Hence, the data is used as is without any record deletion.

#### Gender diversification among the lot:

```python
hr['Sex'].value_counts().plot(kind='bar')
```

#### Department wise gender diversification:

```python
plt.figure(figsize=(16,9))
ax=sns.countplot(x=hr['Department'],hue=hr['Sex'])
```

#### Let us now examine the racial background of the employees in the said dataset:

```python
hr['RaceDesc'].value_counts().plot(kind='pie')
```

#### It can be seen that majority of the employees have a 'White' ethnicity. Let us put a number on that by means of a percentage. We must note that the said organisation is based out of the United States of America.

```python
hr.RaceDesc.name = 'Racial group'
display(hr.RaceDesc.value_counts(),
        hr.RaceDesc.value_counts(normalize=True) * 100)
```

#### According to 2018 data by the U.S Bureau of Labor Statistics, the country's labor force is made of 78% Whites, 13% Blacks and 6% Asians (https://www.bls.gov/opub/reports/race-and-ethnicity/2018/home.htm). Taking that as a baseline, we can say that the company is diverse from a racial standpoint.

#### Let us now see how the relationship status of the employees fare overall:

```python
hr['MaritalDesc'].value_counts().plot(kind='bar')
```

#### As can be seen, most of the employees are 'Single'.

#### The dataset contains information on the results of a survey conducted to determine Employee Satisfaction. This can be visualized in the form of a histogram as shown below:

```python
plt.figure(figsize=(12,6))
sns.countplot(x='EmpSatisfaction',data=hr,saturation=1)
```

#### Performance Score is another parameter that each employee is credited with, based on his/her performance. Each employee is mapped to a Manager who in turn assigns a performance score. This can be seen below, where the graph shows performance scores in each department.

```python
plt.figure(figsize=(12,6))
sns.countplot(data=hr,x='Department',hue='PerfScoreID')
```

#### The ethnic diversity mapped along with the citizenship status of the employees:


```python
plt.figure(figsize=(16,9))
ax=sns.countplot(x=hr['CitizenDesc'],hue=hr['RaceDesc'])
```

#### Employee status marks the current status of employment for each employee. This is classified as: Active, Terminated for cause, Future Start, Leave of absence and Volunatry termination. We can see that most employees are actively employed in the company.

```python
hr['EmploymentStatus'].hist(bins=50)
```

#### There are various positions in the said company. The plot below showcsaes the count of various positions in the company.

```python
plt.figure(figsize=(16,9))
hr['Position'].value_counts().plot(kind='bar')
```

#### Now, let us dive into the salary and payscale of the employees. The plot shows which positions have highest pay rates.

```python
plt.figure(figsize=(16,9))
hr.groupby('Position')['PayRate'].sum().plot(kind='bar')
```

#### The mean, median and atandard deviations of the pay rates are as follows: 

```python
hr['PayRate'].describe()
```

#### Department wise pay rates:

```python
hr.groupby('Department')['PayRate'].sum().plot(kind='bar')
```

#### The overall correlation of various parameters can be seen in a heatmap as shown below. Note that only relevant IDs with numeric datatypes have been taken into consideration.

```python
plt.figure(figsize=(16,16))
sns.heatmap(hr.corr(),annot=True)
```
